```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Data

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

# MATLAB Method

There are two ways to create scatter plots using the MATLAB approach:

- plt.plot()
- plt.scatter()

- https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html
- https://matplotlib.org/stable/gallery/shapes_and_collections/scatter.html

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(x_axis, y_axis)
```

```python
plt.scatter(x_axis, y_axis, color="red", label="Chicago")
plt.legend()
```

# Bubble Chart MATLAB Method

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

## List comprehension

- new_list = [expression for item in list if conditional]
- https://docs.python.org/3.6/tutorial/datastructures.html#list-comprehensions

```python
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

# Create a Scatter Plot Using the Object-Oriented Interface



```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s= [i * 5 for i in y_axis], color = "skyblue",alpha = 0.8, edgecolors = "black",linewidths=2, label = "Boston")
ax.legend()

ax.set_title("PyBer Fare bt Month")
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Months')

ax.set_xlim(5,50)

ax.invert_yaxis()  # labels read top-to-bottom
```
