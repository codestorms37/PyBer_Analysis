```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

Let's break down what the code is doing:

1. The plt.subplots() function returns a tuple that contains a figure and axes object(s).
2. fig, ax = plt.subplots() unpacks the tuple into the variables fig and ax.
3. The first variable, fig, references the figure.
4. The second variable, ax, references the axes of the figure and allows us to annotate the figure.
5. The ax.plot(x_axis, y_axis) plots the data from the x and y axes.

```python
# The fig, ax = plt.subplots() line of code can also be written like this.
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
```

Where fig references the figure, and ax references the axes of the figure and allows us to annotate the figure.

```python
fig = plt.figure() #same as above
ax = fig.add_subplot() #same as above
ax.plot(x_axis, y_axis) # New

plt.show()
```

```python
# To create a quick graph using the object-oriented interface method, use the following two lines:
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)

plt.show()
```

# plt.show()

The plt.show() function looks for all the active figure objects and opens a window that displays them if you have two or more sets of data to plot.

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# marker, color, linewidth, grid

# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()

plt.show()
```

```python
# Matplotlib Object-Oriented Functions
w = 20
h = 10
x = x_axis
y = y_axis
y_min = 5
y_max = 45
x_min = -1
x_max = 12
my_label = "Boston"

# Change the size of the figure in pixels. Add this in the subplots() function.
fig, ax = plt.subplots(figsize=(w, h))

# Add a label that will be added to the legend.
ax.plot(x, y, label=my_label, color="g", linewidth=5, marker = "d", markersize = 20)

# Sets the min and max range of the y-axis.
ax.set_ylim(y_min, y_max)

# Sets the min and max range of the x-axis.
ax.set_xlim(x_min, x_max)

# Add a label to the x-axis.
ax.set_xlabel('x label')

# Add a label to the y-axis.
ax.set_ylabel('y label')

# Add a title.
ax.set_title("Title")

# Add a legend.
ax.legend()

# Add a grid to the chart.
ax.grid()

# Saves the figure with the given extension. Added at the end of your script.
plt.savefig("resources\practice_plot.png")
```
