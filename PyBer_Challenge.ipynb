```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
# File to Load (Remember to change these)
city_data_to_load = os.path.join(".","resources","city_data.csv")
ride_data_to_load = os.path.join(".","resources","ride_data.csv")

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
group_city_type = pyber_data_df.groupby(by="type")

group_city_data_by_city_type = city_data_df.groupby("type")
```

```python
#  1. Get the total rides for each city type
rides_by_city_type = group_city_type.count()[["ride_id"]]
rides_by_city_type
```

```python
# 2. Get the total drivers for each city type
drivers_by_city_type = group_city_data_by_city_type.sum()[["driver_count"]]
drivers_by_city_type
```

```python
#  3. Get the total amount of fares for each city type
total_fares_by_city_type = group_city_type.sum()[["fare"]]
total_fares_by_city_type
```

```python
#  4. Get the average fare per ride for each city type. 
avg_fare_by_city_type = group_city_type.mean()[["fare"]]
avg_fare_by_city_type
```

```python
# 5. Get the average fare per driver for each city type. 
avg_fare_per_driver_by_city_type = total_fares_by_city_type["fare"] / drivers_by_city_type['driver_count']
avg_fare_per_driver_by_city_type
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = rides_by_city_type
pyber_summary_df["Total Rides"] = rides_by_city_type["ride_id"]
pyber_summary_df["Total Drivers"] = drivers_by_city_type
pyber_summary_df["Total Fares"] = total_fares_by_city_type
pyber_summary_df["Average Fare per Ride"] = avg_fare_by_city_type
pyber_summary_df["Average Fare per Driver"] = avg_fare_per_driver_by_city_type

pyber_summary_df = pyber_summary_df.drop(columns="ride_id")

pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
pyber_summary_df
```

```python
#  8. Format the columns.
pyber_summary_df_formated = pyber_summary_df.copy()

pyber_summary_df_formated["Total Rides"] = pyber_summary_df["Total Rides"].map("{:,.0f}".format)
pyber_summary_df_formated["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:,.0f}".format)
pyber_summary_df_formated["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df_formated["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:,.2f}".format)
pyber_summary_df_formated["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:,.2f}".format)


pyber_summary_df_formated
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
groupby_type_and_date = pyber_data_df.groupby(by=["type","date"])

total_fare_by_type_and_date_df = groupby_type_and_date.sum()[["fare"]]
total_fare_by_type_and_date_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
total_fare_by_type_and_date_df = total_fare_by_type_and_date_df.reset_index()
total_fare_by_type_and_date_df
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
fare_pivot = total_fare_by_type_and_date_df.pivot(index="date", columns="type", values="fare")
fare_pivot
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
fare_Jan_Apr = fare_pivot.loc['2019-01-01':'2019-04-29']
fare_Jan_Apr
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)

#fare_Jan_Apr=fare_Jan_Apr.reset_index()
#fare_Jan_Apr.dtypes

fare_Jan_Apr.info()
```

```python
fare_Jan_Apr.index = pd.to_datetime(fare_Jan_Apr.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
fare_Jan_Apr.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
weekly_fares_df = fare_Jan_Apr.resample('W').sum()
weekly_fares_df.head(10)
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

weekly_fares_plot = weekly_fares_df.plot(figsize = (20,6))

# Add a title 
weekly_fares_plot.set_title("Number of Rides by Gender")
#  Add a x- and y-axis label.
weekly_fares_plot.set_xlabel("Month")
weekly_fares_plot.set_ylabel("Total Fare ($USD)")


target_path = os.path.join(".","analysis","PyBer_fare_summary.png")

plt.savefig(target_path)

plt.show()
```
