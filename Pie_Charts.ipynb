```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Data

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# Explode Values
explode_values = (0, 0.5, 0, 0.5, 0, 0.5, 0, 0.5, 0, 0.5, 0, 0.5)
```

# MATLAB Method

- https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html


```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8)) ## Ths has to go always first (or at least before plt.pie or something like that)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

# Colors

- https://matplotlib.org/stable/gallery/color/named_colors.html


```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(
    y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%',
    counterclock = False
)

plt.show()
```

# Object-Oriented Interface

- https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.pie.html

```python
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

mycolors = ["#A50000","#B50001","#C50002","#D50003","#E50004","#F50005","#150006","#250007","#350008","#450009","#55000A","#650000"]

fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(
    x=y_axis,
    explode=explode_values,
    labels=x_axis,
    autopct='%.1f%%',
    pctdistance= 0.9,
    counterclock = False,
    shadow = True,
    startangle = 90,
    colors = mycolors
)
plt.show()
```
