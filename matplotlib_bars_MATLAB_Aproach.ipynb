```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

# Quick and dirty

```python
plt.bar(x_axis, y_axis)
```

# The basic Titles with label / legend

```python
# Create the plot

plt.bar(x_axis, y_axis, label="Boston")

# Create labels for the x and y axes.
plt.xlabel("\n Date \n")
plt.ylabel("\n Fare($) \n")

# Create a title.
plt.title("\n PyBer Fare by Month \n")

# Add the legend.
plt.legend()

plt.show()
```

```python
# Create the plot
f, ax = plt.subplots(figsize=(15,5))
plt.bar(x_axis, y_axis, color="green", edgecolor = "orange",linewidth=5, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("\n Date \n",fontsize = 20)
plt.ylabel("\n Fare($) \n",fontsize = 20)
# Create a title.
plt.title("\n PyBer Fare by Month \n", fontsize = 30)
# Add the legend.
plt.legend(fontsize = 20)

# the same as above
# ax.legend(fontsize = 14)


plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)

plt.show()
```

```python
# Create the plot
plt.barh(x_axis, y_axis)

# invert axis
plt.gca().invert_yaxis()
```

```python
# Create the plot
plt.barh(y_axis, x_axis)
```

```python
# Create the plot
plt.barh(x_axis, y_axis, color="magenta", label = "Boston")

plt.legend()

# invert axis
plt.gca().invert_yaxis()

# Create labels for the x and y axes.
plt.ylabel("\n Date \n")
plt.xlabel("\n Fare($) \n")

# Create a title.
plt.title("\n PyBer Fare by Month \n")
```
